```python
def to_do_list():
    tasks = [] # Changed 'task' to 'tasks'
    while True:
        print("1. Add task")
        print("2.Remove task")
        print("3. Show task")
        print("4.Quit")
        choice = input("Enter your choice:") # Fixed typo 'choise' to 'choice'
        if choice == "1":
            task_item = input("Enter your task:") # Renamed local 'task' variable to avoid confusion
            tasks.append(task_item)
        elif choice == "2":
            task_item = input("Enter task to remove:") # Renamed local 'task' variable
            if task_item in tasks:
                tasks.remove(task_item)
            else:
                print("Task not found")
        elif choice == "3":
            print("Your tasks:")
            if not tasks: # Added check for empty list
                print("No tasks yet.")
            for task_item in tasks: # Renamed loop variable
                print("- " + task_item)
        elif choice == "4":
            print("Goodbye!") # Removed ':'
            break # Added break to exit loop
        else:
            print("Invalid choice please re enter") # Fixed typo 'choise' to 'choice'
to_do_list()
```
